```python
df= pd.read_csv("fixed_running_data.csv", parse_dates=[0])
```

```python
cols= ["Date", "Miles", "Time"]
df.columns= cols
```

```python
df.head()
```

```python
df.rename(columns={df.columns[0]:"dates"})
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
df.plot()
```

```python
raw_time_fmt= df.Time[0]

def get_total_seconds(raw_time_fmt):
    if raw_time_fmt is np.nan:
        return np.nan
    hrs, mins, seconds= str(raw_time_fmt).split(":")
    seconds= int(seconds) + 60 * int(mins) + 60 * 60 * int(hrs)
    return seconds 
```

```python
df['Seconds']= df.Time.map(get_total_seconds)
```

```python
df['Minutes']= df['Seconds'].map(lambda x: x/60)
```

```python
df['Min_per_mile']= df['Minutes']/df['Miles']
```
